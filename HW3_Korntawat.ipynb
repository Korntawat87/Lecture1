# 6.) The company you are working for wants to target at a False Positive rate of 5% what threshold should you use? (Use oversampled data)

```python
# ASK chatgpt
# If you dont like the code ask in a different way
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
clf = LogisticRegression()
clf.fit(x_train,y_train)

probs = clf.predict_proba(x_test)

#Find the treshold that give a False Positive rate of 5%
threshold = 0
for i in np.arange(0, 1, 0.01):
    preds = probs[:,1] > i
    cm = confusion_matrix(y_test, preds)
    fpr = cm[0,1] / (cm[0,0] + cm[0,1])
    if fpr <= 0.05:
       threshold = i
       break
```

```python
threshold
```

# 7.) If the company makes .02*amt on True transactions and loses -amt on False (Use original data)

```python
preds = probs[:,1]>0.01
```

```python
x_test
```

```python
x_test['Preds'] = preds
```

```python
x_test['signal'] = x_test['Preds'] * y_test
```

```python
x_test
```

```python
comp_df = pd.concat([x_test, y_test], axis =1)
```

```python
comp_df['Preds']= [0 if i == False else 1 for i in comp_df['Preds']]
```

```python
comp_df['FN'] = (comp_df['is_fraud']==1) & (comp_df['Preds']==0)
```

```python
comp_df
```

```python
comp_df['profit'] = comp_df['signal'] * 0.02 * comp_df['amt'] + comp_df['FN'] * -comp_df['amt'] 
```
